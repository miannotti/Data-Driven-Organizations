```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance.csv") #this is a comment
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(19,5))
sns.histplot(df["charges"], bins=30, color="skyblue")
plt.title("Distribution of charges")
```

```python
plt.figure(figsize=(19,5))
sns.boxplot(x="smoker", y="charges", data=df, color="skyblue", hue="sex")
plt.title("Charges by smoker")
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(x="bmi", y="charges", data=df, color="skyblue")
plt.title("Charges by smoker")
```

```python
plt.figure(figsize=(19,5))
sns.violinplot(x="region", y="charges", data=df, palette="Set2", hue="region")
```

```python
plt.figure(figsize=(19,5))
sns.violinplot(x="region", y="charges", data=df, palette="Set2", hue="smoker")
```

```python
plt.figure(figsize=(19,5))
sns.kdeplot(data=df, x="charges", hue="smoker")
```

```python
plt.figure(figsize=(19,5))
sns.scatterplot(x="age", y="charges", data=df, color="green")
```

```python
plt.figure(figsize=(19,5))
sns.scatterplot(x="age", y="charges", data=df, color="green", hue="smoker")
```

```python
plt.figure(figsize=(19,5))
sns.scatterplot(x="bmi", y="charges", data=df, color="green", hue="smoker")
```

```python
plt.figure(figsize=(19,5))
corr = df.corr(numeric_only=True)
```

```python
corr
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
```

```python
plt.figure(figsize=(19,5))

df['smoker_num'] = df['smoker'].map({'yes':1, 'no':0})
df['sex_num'] = df['sex'].map({'male':1, 'female':0})

cols = ['age', 'bmi', 'children', 'charges', 'smoker_num', 'sex_num']
corr = df[cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Key Relationships")
plt.show()
```
